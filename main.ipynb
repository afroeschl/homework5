$$
\begin{aligned}
y(t) &= -\frac{g}{2} t^2 + v_0\sin(\theta) t + y_0 \\
x(t) &= v_0\cos(\theta)t
\end{aligned}
$$

$$
y(x) = f(x) = - \frac{g}{2 v_0^2 \cos^2(\theta)} x^2 + \tan(\theta) x
\tag{1}
$$

$$
d = \frac{v_0^2\sin(2\theta)}{g}

$$

- $g$ ist die Erdbeschleunigung $(\text{9.81 m/s}^2)$.
